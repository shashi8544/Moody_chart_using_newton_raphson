```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from matplotlib.pyplot import figure
from math import log10
from math import sqrt
from math import pow
import matplotlib.ticker
from sympy import *
f1=[]
Re1=[]
e_d=[0.00001,0.00005,0.0001,0.0002,0.0004,0.0006,0.0008,0.001,0.002,0.004,0.006,0.008,0.01,0.015,0.02,0.03,0.04,0.05]
for Rug in e_d:  
    for Re in range(2300,10**8,10**7):
        Range=range(2300,10**8,10**7)
        def newtonraphson(func, funcderiv, x,n):
            def f(x):
              f =eval(func)
              return f

            def df(x):
              df = eval(funcderiv)
              return df
 
            for itretion in range(1,n):
              i=x-(f(x)/df(x))

              x = i
              f=1/(x*x)
            f1.append(f)
            Re1.append(Re)

        newtonraphson("x-1.14+2.0*(log10(Rug+9.3*x/(Re)))", "(x+(2*9.3/(Re*(Rug+(9.3/Re)))))", 3.59,2)      

l1=len(f1) 
l2=len(Range) 
l3=l1/l2
l3=round(l3) 


flam=[]
RElam=[]
for Re_lam in range(700,2300,50):
    flaminar=64/Re_lam 
    flaminar=np.around(flaminar,decimals=3)  
    RElam.append(Re_lam)
    flam.append(flaminar)  
plt.figure(figsize=(13,5))
plt.plot(RElam,flam,'r-',markersize=2, linewidth=2)
for i in range(0,l3,1): 
    Re1list=[Re1[i:i+l2] for i in range(0, len(Re1), l2)]
    f1list=[f1[i:i+l2] for i in range(0, len(f1), l2)] 
    f1list_=f1list[i]
    p=f1list_[l2-1] 
    plt.figure(figsize=(13,5))
    plt.plot(Re1list[i],f1list[i],'-',markersize=2, linewidth=2)
    plt.title('Moody Diagram in log scale',size='xx-large')
    plt.annotate(e_d[i], (1.2*10**8,p),size=10)   
    plt.xlabel('Reynolds Number,Re',size=18)
    plt.xscale('log')
    plt.yscale('log')
    plt.ylabel('Friction Factor,f',size=18)
    plt.text(1.1, 0.5, 'Relative Roughness,e/D', {'color': 'k', 'fontsize': 15},    
         horizontalalignment='left',verticalalignment='center', rotation=270, clip_on=False,
         transform=plt.gca().transAxes) 
    plt.axvspan(2300, 4000, facecolor='gainsboro', alpha=0.5) 
    plt.grid(which='major', linestyle='--', linewidth='1.0', color='k') 
    plt.grid(which='minor', linestyle=':', linewidth='0.5', color='k') 
    ax = plt.gca()
    ax.yaxis.set_major_formatter(matplotlib.ticker.ScalarFormatter())
    ax.yaxis.set_minor_formatter(matplotlib.ticker.ScalarFormatter())  
    plt.autoscale(enable=True, axis='y')
    plt.minorticks_on


plt.annotate('Transition Zone\n 2,300 < Re < 4,000', (2900,0.05),fontsize=12)
plt.annotate('Laminar Flow\nf=64/Re', (450,0.04),fontsize=12)
```
